```python
import librosa
import numpy as np

# Define the path to the input audio file
input_audio = '/content/Music/001.mpeg' # Changed to a single file path

# Load the audio file
y, sr = librosa.load(input_audio)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Print the shape of the extracted features
print("Shape of MFCC features:", mfccs.shape)
```
